```python
#Importing required libraries fro visualisation
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

**Pairplot** is a data visualization technique that is used to explore the relationships between multiple variables in a dataset. It is a scatterplot matrix, which means that it creates a separate scatterplot for each pair of variables in the dataset. This allows you to quickly and easily see how different variables are related to each other.

```python
#Importing dataset (Daily Activity)
df=pd.read_csv('Daily Activity.csv')
```

```python
#Plotting Pair-Plot
sns.pairplot(df)
plt.show()
```
